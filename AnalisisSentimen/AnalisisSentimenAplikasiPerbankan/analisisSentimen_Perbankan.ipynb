## 2. Preprocessing

```python
# Fungsi untuk memfilter data berdasarkan panjang kata, jadi kata yang muncul minimal 3 kata dan maksimal 50 kata, jika tidak memenuhi kriteria tersebut maka data tersebut akan dihapus
def filter_tokens_by_length(dataframe, column, min_words, max_words):
    words_count = dataframe[column].astype(str).apply(lambda x: len(x.split()))
    mask =  (words_count >= min_words) & (words_count <= max_words)
    filtered_df = dataframe[mask]
    return

min_words = 3
max_words = 50
df = filter_tokens_by_length(df, 'content', min_words, max_words)
```

```python
# Normalisasi
norm = {
    ' bisa2 ': ' bisa ', 
    ' kmn ': ' kemana ', 
    ' kmna ': ' kemana ', 
    ' ngga ': ' tidak ', 
    ' nggak': ' tidak ', 
    ' tdk ': ' tidak ', 
    ' gk ': ' tidak ', 
    ' ga ': ' tidak ',
    ' yg ': ' yang ',
    ' wkwk ': ' ',
    ' udh ': ' udah ',
    ' apk ': ' aplikasi ', 
    ' tiktokan ': ' tiktok ', 
    ' proficional ': ' profesional ',
    ' kalo ': ' kalau ',
    ' uwang ': ' uang ',
    ' skrng ': ' sekarang '
}

def normalisasi(str_text):
    for i in norm:
        str_text = str_text.replace(i, norm[i])
    return str_text

df['content'] = df['content'].astype(str).apply(lambda x: normalisasi(x))
```

```python
# Stopwords
import Sastrawi
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory, StopWordRemover, ArrayDictionary

# tujuan dari stopword adalah untuk membuang kata kata yang tidak memiliki bobot, contohnya kata di, yang, dengan, dan lain lain
more_stop_words = []

stop_words = StopWordRemoverFactory().get_stop_words()
stop_words.extend(more_stop_words)

new_array = ArrayDictionary(stop_words)
stop_words_remover_new = StopWordRemover(new_array)

def stopword(str_text):
    str_text = stop_words_remover_new.remove(str_text)
    return str_text

df['content'] = df['content'].apply(lambda x: stopword(x))
df.sample(5)
```

```python
# Visualisasi untuk melihat kata kata terbanyak untuk masing masing perbankan

def hitung_kata_terbanyak(df, aplikasi, sentimen):
    komentar = df[(df['application'] == aplikasi) & (df['sentimen'] == sentimen)]['content']
    kata_counter = Counter()
    komentar.apply(lambda x: kata_counter.update(x.split()))
    return pd.DataFrame(kata_counter.most_common(10), columns=['Kata', 'Frekuensi'])


def visualisasi_kata(top_kata_df, judul, warna, posisi):
    plt.subplot(3, 3, posisi)
    plt.bar(top_kata_df['Kata'], top_kata_df['Frekuensi'], color=warna)
    plt.title(judul)
    plt.xticks(rotation=45)

aplikasi_list = ['BNI', 'BRI', 'BCA']
sentimen_list = ['Negatif', 'Positif', 'Netral']

palette = {
    'BNI': sns.color_palette("Reds", 3),
    'BRI': sns.color_palette("Greens", 3),
    'BCA': sns.color_palette("Blues", 3)
}

plt.figure(figsize=(20, 15))
posisi = 1

for aplikasi in aplikasi_list:
    for idx, sentimen in enumerate(sentimen_list):
        top_kata_df = hitung_kata_terbanyak(df, aplikasi, sentimen)
        warna = palette[aplikasi][idx]
        judul = f'{aplikasi} - 10 kata {sentimen} terbanyak'
        visualisasi_kata(top_kata_df, judul, warna, posisi)
        posisi += 1

plt.tight_layout()
plt.show()
```

## 5. Prepare data training

```python
# Memisahkan data berdasarkan aplikasi
df_bca = df[df['application'] == 'BCA']
df_bni = df[df['application'] == 'BNI']
df_bri = df[df['application'] == 'BRI']
```

```python
# Variabel untuk memisahkan data training dan juga data testing

X_bca = df_bca['content']
y_bca = df_bca['sentimen']

X_bni = df_bni['content']
y_bni = df_bni['sentimen']

X_bri = df_bri['content']
y_bri = df_bri['sentimen']
```

```python
# data ini akan digunakan untuk training data menggunakan 4 algoritma machine learning (Random Forest, KNN, SVM, dan Naive Bayes)
data = {
    'BCA': (X_bca, y_bca),
    'BNI': (X_bni, y_bni),
    'BRI': (X_bri, y_bri)
}
```

## 6. Training data

```python
# Training data

# Inisialisasi list kosong untuk menyimpan hasil evaluasi model
results = []

# Loop untuk setiap aplikasi dan data latihannya (X, y)
for app, (X, y) in data.items():
    # Membagi data menjadi data latih (85%) dan data uji (15%)
    x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)

    # Membuat instance CountVectorizer untuk mengonversi data teks menjadi matriks token
    vectorizer = CountVectorizer()

    # Melatih vectorizer pada data latih dan mengubahnya menjadi matriks fitur
    x_train_vectorized = vectorizer.fit_transform(x_train).toarray()

    # Mengubah data uji menjadi ruang fitur yang sama seperti data latih
    x_test_vectorized = vectorizer.transform(x_test).toarray()

    # Mendefinisikan dictionary yang berisi berbagai classifier untuk pelatihan dan evaluasi
    classifiers = {
        'KNN': KNeighborsClassifier(n_neighbors=5),  # Klasifikasi K-Nearest Neighbors dengan 5 tetangga
        'RandomForest': RandomForestClassifier(),     # Klasifikasi Random Forest
        'NaiveBayes': MultinomialNB(),                # Klasifikasi Naive Bayes Multinomial
        'SVM': SVC(kernel='linear'),                  # Support Vector Machine dengan kernel linear
    }

    # Loop untuk setiap classifier dan melatihnya pada data latih
    for clf_name, clf in classifiers.items():
        # Melatih classifier menggunakan data latih
        clf.fit(x_train_vectorized, y_train)

        # Memprediksi label untuk data uji menggunakan classifier yang sudah dilatih
        y_pred = clf.predict(x_test_vectorized)

        # Menghitung metrik evaluasi untuk classifier
        accuracy = accuracy_score(y_test, y_pred)  # Menghitung akurasi, proporsi sampel yang diklasifikasikan dengan benar
        recall = recall_score(y_test, y_pred, average='weighted')  # Menghitung rata-rata tertimbang recall
        precision = precision_score(y_test, y_pred, average='weighted')  # Menghitung rata-rata tertimbang precision
        f1 = f1_score(y_test, y_pred, average='weighted')  # Menghitung rata-rata tertimbang F1 score

        # Menyimpan hasil evaluasi dalam dictionary dan menambahkannya ke dalam list results
        results.append({
            'Application': app,  # Nama aplikasi
            'Classifier': clf_name,  # Nama classifier
            'Accuracy': accuracy,  # Skor akurasi
            'Recall': recall,  # Skor recall
            'Precision': precision,  # Skor precision
            'F1 Score': f1  # Skor F1
        })
```

## 7. Comparative evaluation

```python
# Menyimpan data evaluasi ke dalam variabel df_results
df_results = pd.DataFrame(results)
df_results
```

```python
# Visualisasi Confusion Matrix

# Memilih classifier terbaik berdasarkan akurasi untuk setiap aplikasi
best_classifiers = df_results.loc[df_results.groupby('Application')['Accuracy'].idxmax()]

# Loop untuk setiap baris dalam data classifier terbaik
for index, row in best_classifiers.iterrows():
    # Mendapatkan nama aplikasi dan classifier terbaik
    app = row['Application']
    clf_name = row['Classifier']
    
    # Membagi data aplikasi menjadi data latih dan uji (85% latih, 15% uji)
    x_train, x_test, y_train, y_test = train_test_split(data[app][0], data[app][1], test_size=0.15, random_state=42)
    
    # Membuat dan melatih CountVectorizer pada data latih untuk mengonversi teks menjadi fitur numerik
    vectorizer = CountVectorizer()
    x_train_vectorized = vectorizer.fit_transform(x_train).toarray()  # Konversi data latih ke array numerik
    x_test_vectorized = vectorizer.transform(x_test).toarray()  # Konversi data uji ke array numerik menggunakan vektor yang sama
    
    # Memuat classifier terbaik dari dictionary classifiers
    clf = classifiers[clf_name]
    
    # Melatih classifier pada data latih
    clf.fit(x_train_vectorized, y_train)
    
    # Memprediksi label untuk data uji
    y_pred = clf.predict(x_test_vectorized)
    
    # Membuat confusion matrix dari label sebenarnya (y_test) dan prediksi (y_pred)
    cm = confusion_matrix(y_test, y_pred)
    
    # Membuat visualisasi confusion matrix menggunakan heatmap
    plt.figure(figsize=(8, 6))  # Menentukan ukuran figure
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',  # Menampilkan nilai dalam setiap sel heatmap
                xticklabels=np.unique(y_train),  # Menentukan label sumbu x (prediksi)
                yticklabels=np.unique(y_train))  # Menentukan label sumbu y (label sebenarnya)
    
    # Menambahkan judul dan label sumbu untuk visualisasi
    plt.title(f'Confusion Matrix for {app} - {clf_name}')
    plt.xlabel('Predicted Label')  # Label untuk sumbu x
    plt.ylabel('True Label')  # Label untuk sumbu y
    
    # Menampilkan heatmap
    plt.show()
```

```python
# Mengatur tema tampilan grafik menggunakan seaborn
sns.set(style='whitegrid')  # Menggunakan gaya 'whitegrid' agar grafik terlihat bersih dengan grid putih
palette = sns.color_palette('Set2')  # Mengatur palet warna 'Set2' untuk memberikan variasi warna pada grafik

# Membuat figure dengan 3 subplot secara horizontal, ukuran keseluruhan 18x6 inci
fig, axes = plt.subplots(1, 3, figsize=(18, 6))  

# Mendapatkan daftar unik nama aplikasi dari dataframe hasil (df_results)
applications = df_results['Application'].unique()  

# Looping untuk setiap aplikasi dengan memasangkan axis (ax) ke subplot
for ax, app in zip(axes, applications):
    # Memfilter hasil evaluasi hanya untuk aplikasi tertentu
    app_results = df_results[df_results['Application'] == app]  
    
    # Mengubah dataframe ke format 'long' agar dapat digunakan dengan seaborn
    app_results_melted = app_results.melt(
        id_vars=["Application", "Classifier"],  # Kolom tetap yang tidak diubah
        value_vars=["Accuracy", "Recall", "Precision", "F1 Score"],  # Kolom metrik yang akan diubah menjadi baris
        var_name="Metric",  # Nama kolom baru untuk jenis metrik
        value_name="Score"  # Nama kolom baru untuk nilai metrik
    )
    
    # Membuat grafik batang (bar plot) untuk setiap aplikasi
    sns.barplot(
        data=app_results_melted,  # Data yang digunakan
        x='Score',  # Sumbu x: nilai metrik
        y='Classifier',  # Sumbu y: nama classifier
        hue='Metric',  # Warna berdasarkan jenis metrik
        ax=ax,  # Menentukan subplot yang digunakan
        ci=None,  # Tidak menggunakan interval kepercayaan
        palette=palette  # Menggunakan palet warna yang sudah diatur
    )
    
    # Menambahkan judul pada setiap subplot
    ax.set_title(f'Comparison of Evaluation Metrics for {app}', fontsize=16, weight='bold')
    # Menambahkan label pada sumbu x
    ax.set_xlabel('Score', fontsize=14)
    # Menambahkan label pada sumbu y
    ax.set_ylabel('Classifier', fontsize=14)
    # Membatasi nilai maksimum pada sumbu x hingga 1 (karena skor maksimal adalah 1)
    ax.set_xlim(0, 1)
    # Mengatur rotasi label sumbu y agar tetap horizontal
    ax.tick_params(axis='y', rotation=0)

    # Menambahkan grid horizontal pada sumbu x
    ax.xaxis.grid(True, linestyle='--', alpha=0.7)

    # Looping untuk setiap batang dalam bar plot
    for p in ax.patches:
        # Mendapatkan nilai skor dari panjang batang
        score = p.get_width()
        # Menambahkan anotasi pada batang dengan nilai skornya
        ax.annotate(
            f'{score:.2f}',  # Format nilai skor hingga dua desimal
            (score, p.get_y() + p.get_height() / 2),  # Posisi anotasi
            ha='center', va='center',  # Penyelarasan horizontal dan vertikal
            color='black', fontsize=10,  # Warna dan ukuran font
            xytext=(5, 0),  # Posisi offset untuk teks anotasi
            textcoords='offset points',  # Menggunakan sistem koordinat offset
            bbox=dict(facecolor='white', edgecolor='none', boxstyle='round,pad=0.3')  # Menambahkan latar belakang putih pada teks
        )

# Mengatur tata letak agar grafik tidak saling tumpang tindih
plt.tight_layout()

# Menampilkan grafik
plt.show()
```
